### Test leakage
What happened:

1. We trained on `train`, tuned the threshold on `val`, evaluated on `test`
2. We looked at `TEST`-misclassifications, corrected false labels in the dataset and retrained the model with information we got from the test set
3. Images updated after first run: 24 out of ~10000 swapped image labels  -> 2.4% test leakage
---
for the next runs we will only look at false predictions inside the `train` and `val` set...
